```python
def visualize(filename): # create a function called visualize
  "Plot the visualizations"
  # create a pandas dataframe
  df = pd.read_csv(filepath_or_buffer = filename, header = None)
  # count number of columns in the df
  total_columns = len(df.axes[1]) # Axes of 0 is for a column
  #print("Number of Columns in",datasets,": ", total_columns) # for debugging
  
  # for each column, we check get its average, max and min
  # to do this we create two arrays(list) that will have column numbers(0-39) and their respective stats
  
  features_list = list(range(0, 40)) # 0 - 39
  #print("Features list: ",features_list) #for debugging

  # we now get average of a column, put in array
  average = [] # list to hold average values of columns
  max = [] # list to hold maximum values of columns
  min = [] # list to hold minimum values of columns

  for count in range(0,total_columns):
    col_list = df.iloc[:, count].values.tolist() 
    # get the statistics
    column_average = np.mean(col_list) # get average of current column
    column_max = np.max(col_list) # get maximum value of current column
    column_min = np.min(col_list) # get minimum value of current column
    #print(column_average) # for debugging

    # we now put the values in their lists
    average.append(column_average) 
    max.append(column_max) 
    min.append(column_min) 


  #print("Features average: ",average) #for debugging

  "PLOT AVERAGE OF FEATURES"
  # we now plot a line chart of columns and their averages
  plt.plot(features_list, average, 'yellow', label = 'Avg')
  # Add title and axis names
  title = "Dataset #"+str(n)
  plt.title(title)
  plt.xlabel('Feature#')
  plt.ylabel('Values')

  "PLOT MAX OF FEATURES"
  plt.plot(features_list, max, 'green', label = 'Max')

  "PLOT MIN OF FEATURES"
  plt.plot(features_list, min, 'orange', label = 'Min')

  # add horizontal lines in graphs
  axes = plt.gca()
  axes.yaxis.grid()

  # show the plots
  plt.legend()
  plt.show()
```
